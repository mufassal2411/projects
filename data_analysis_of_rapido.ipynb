```python
buddy = buddy_copy.copy()
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
buddy['services'] = label_encoder.fit_transform(buddy['services'])
buddy['ride_status'] = label_encoder.fit_transform(buddy['ride_status'])
```

```python
buddy.head()
```

```python
buddy.info()
```

```python
cols = ['services','ride_status','duration','distance']
x=buddy[cols]
y=buddy['total_fare']
```

```python
x.head()
```

```python
y.head()
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape ,x_test.shape,y_train.shape ,y_test.shape
```

```python
x_train.head()
```

```python
#training model
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
#prediction
y_pred=lr.predict(x_test)
y_pred
```

```python
#evaluation
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,y_pred)
score
```

```python
#support vector regresssion
from sklearn.svm import SVR
svr = SVR()
svr.fit(x_train,y_train)
```

```python
#prediction
y_pred = svr.predict(x_test)
y_pred
```

```python
#evaluate
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
#decision tree model
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()

# fit the regressor with X and Y data
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
#random forest method
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
#predict
y_pred = model.predict(x_test)
y_pred
```

```python
#evaulate
mse = mean_squared_error(y_test, y_pred)
mse
```
